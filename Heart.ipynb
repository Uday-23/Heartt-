```python
#READING THE DATASET
import pandas as pd
df=pd.read_csv("/home/udzi/Downloads/heart.csv")
df
```

```python
#CHECKING IF THERE ARE ANY NULL VALUES IN THE DATAFRAME
df.isna().sum()
```

```python

Y=df["target"] #DATAFRAME WITH THE TARGET COLUMN
X=df.drop("target",axis="columns")#DATAFRAME WITH THE FEATURE COLUMNS
```

```python
X
```

```python
Y
```

```python
#ONE HOT ENCODING OF THE ARRAY USING to_categorical
from tensorflow.keras.utils import to_categorical
Y=to_categorical(Y)
```

```python
#CHECKING THE DIMENSIONS OF THE DATAFRAMES
Y.shape
```

```python
X.shape
```

```python
#SETTING UP THE NECESSARY ARCHITECTURE TO PREDICT THE TARGET COLUMN (STEP1)
from keras.layers import Dense  #IMPORTING DENSE LAYER
from keras.models import Sequential #IMPORTING SEQUENTIAL MODEL
model=Sequential()
model.add(Dense(104,activation="relu",input_shape=(13,)))
model.add(Dense(2,activation="softmax"))
```

```python
#COMPILING THE MODEL (STEP 2)
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
#TRAINING  THE MODEL (STEP 3)
model.fit(X,Y,epochs=10)
```

```python
#SAMPLE DATASET TO TEST THE MODEL
data=pd.DataFrame({"age":[50],
                  "sex":[1],
                   "cp":[2],
                   "trestbps":[150],
                   "chol":[250],
                   "fbs":[1],
                   "restecg":[0],
                   "thalach":[170],
                   "exang":[1],
                   "oldpeak":[2.9],
                   "slope":[2],
                   "ca":[1],
                  "thal":[1]})
```

```python
#PREDICITNG THE VALUES (STEP 4)
model.predict(data)
```
